##### 1. Buatlah model klasifikasi dengan menggunakan SVM untuk data suara, voice.csv. 

Import Library dan Load Data

```python
# import library
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv('../data/voice.csv', encoding='latin-1') # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8

df.head()
```

Menghitung Jumlah tiap Value pada Variabel Label

```python
df['label'].value_counts()
```

Melakukan Encode Variabel Label

```python
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['label'])
df['label'].head()
```

Visualisasi Heatmap

```python
plt.figure(figsize=(20, 18))
corr_matrix = df.corr() 
sns.set(font_scale=1.2) 
sns.heatmap(corr_matrix, 
            cmap='viridis',
            linewidths=1, 
            fmt=".2f",
            annot=True,
            cbar=True, 
            square=True) 

plt.title('Heatmap Korelasi') 
plt.show()
```

Identifikasi Variabel Independen dan Dependen

```python
X = df.iloc[:,:20]
y = df.label
```

Split Data Training dan Testing menjadi 70:30

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 1)
```

```python
x1 = type(X.to_numpy())
x1
```

Import Model SVM

```python
from sklearn.svm import SVC

# Inisialisasi model SVC
svc_model = SVC(kernel='linear', C=1.0, gamma='scale')
```

Melatih Model dengan Data Training dan Prediksi dengan Data Testing

```python
# Latih model dengan data pelatihan
svc_model.fit(X_train, y_train)

# Lakukan prediksi pada data uji
y_pred = svc_model.predict(X_test)
```

Akurasi Model SVM Terhadap Label Sebenarnya dan Hasil Prediksi

```python
from sklearn.metrics import accuracy_score

# Hitung akurasi
accuracy = accuracy_score(y_test, y_pred)

# Cetak hasil akurasi
print(f'Skor Akurasi : {int(accuracy * 100)}%')
```

##### 2. Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,

1. Menggunakan data spam.csv

Load Data

```python
df = pd.read_csv('../data/spam.csv', encoding='latin-1') # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8

df.head()
```

Drop Kolom


```python
# Drop 3 kolom terakhir dengan fungsi iloc
df = df.drop(df.iloc[:,2:], axis=1)

# Cek data
df.head()
```

Ubah Nama v1 dan v2

```python
# Data untuk rename kolom v1 dan v2
new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}

# Rename nama kolom v1 dan v2
df = df.rename(columns=new_cols)

# cek data
df.head()
```

Inspeksi Data

```python
# Cek Jumlah Data Per Kelas
print(df['Labels'].value_counts())
print('\n')

# Cek Kelengkapan Data
print(df.info())
print('\n')

# Cek Statistik Deskriptif
print(df.describe())
```

Encode Label

```python
# Data untuk label
new_labels = {
    'spam': 1,
    'ham': 0
}

# Encode label
df['Labels'] = df['Labels'].map(new_labels)

# Cek data
df.head()
```

Mamisahkan Fitur dengan Label

```python
X = df['SMS'].values
y = df['Labels'].values
```

2. Fitur CountVectorizer dengan mengaktifkan stop_words

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

# Split data training dan testing menjadi 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)

# Inisiasi CountVectorizer
bow = CountVectorizer(stop_words='english')

# Fitting dan transform X_train dengan CountVectorizer
X_train = bow.fit_transform(X_train)

# Transform X_test
X_test = bow.transform(X_test)
```

3. Evaluasi Hasil

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data test
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train * 100:2f}%')
print(f'Hasil akurasi data test: {acc_test * 100:2f}%')
```

#### 3. Buatlah model klasfikasi Multinomial Naive Bayes dengan ketentuan,

1. Menggunakan data spam.csv

```python
X = df['SMS'].values
y = df['Labels'].values
```

2. Fitur TF-IDF dengan mengaktifkan stop_words

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Split data training dan testing menjadi 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)

# Inisiasi CountVectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Fitting dan transform X_train dengan TF-IDF
X_train = tfidf.fit_transform(X_train)

# Transform X_test
X_test = tfidf.transform(X_test)
```

3. Evaluasi hasilnya dan bandingkan dengan hasil pada Tugas no 2.

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data test
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train * 100:f}%')
print(f'Hasil akurasi data test: {acc_test * 100:f}%')
```

4. Berikan kesimpulan fitur mana yang terbaik pada kasus data spam.csv

Berdasarkan hasil evaluasi terhadap akurasi pada model klasisfikasi Multinominal Naive Bayes menggunakan fitur CountVectorized dan TF-IDF, memiliki perbedaan hasil akurasi, yakni:

1. CountVectorized:

    Akurasi data training: 99.358974%

    Akurasi data test: 98.325359%

2. TF-IDF:

    Akurasi data training: 98.410256%

    Akurasi data tes: 96.351675%

Dari perbandingan tersebut, dapat ditarik kesimpulan bahwa model dengan CountVectorized memiliki akurasi yang lebih tinggi dalam data training maupun testing. Namun model dengan CountVectorized cenderung lebih overfitting pada data training, sedangkan TF_IDF memiliki akurasi yang lebih rendah pada data test, hal ini menunjukkan bahwa model ini mungkin memilii generalisasi yang lebih baik, meskipun akurasi data test nya lebih rendah.

Oleh karena itu, CountVectorized mungkin menjadi pilihan yang lebih baik karena menghasilkan akurasi yang lebih tinggi. Namun metode ekstraksi fitur tergantung pada dataset dan tugas tertentu, karena hasil akhir nantinya bisa bervariasi untuk dataset yang berbeda. Jadi sebelum memilih fitur model, baiknya diperhatikan tujuan dari ektraksi data. Apabila ingin mengekstrak fitur teks berdasarkan frekuensi kemunculan kata-kata dalam dokumen tanpa memperhatikan bobot kata-kata tersebut, maka bisa memakai CountVectorized, karena fitur ini tidak memperhatikan pentingnya kata-kata dalam dokumen. Sedangkan ketika ingin mengekstrak fitur teks yang memperhitungkan pentingnya kata-kata, maka TF-IDF bisa menjadi pilihan yang sesuai. Karena fitur ini cenderung memiliki generalisasi yang lebih baik, yang bermanfaat dalam klasifikasi teks yang memerlukan penekanan pada *keyword* tertentu.
